```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#畫函數圖形

```python
# prompt: 劃出 y = sin(x)圖形

# x軸的資料點
x = np.linspace(0, 2 * np.pi, 100)

# y軸的資料點
y = np.sin(x)

# 畫圖
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')

# 顯示圖表
plt.show()
```

#作業1 畫出一個函數圖形

```python
import numpy as np
import matplotlib.pyplot as plt

# 定義參數方程式
t = np.linspace(0, 2 * np.pi, 1000)
x = 16 * np.sin(t) ** 3
y = 8 * np.cos(t) - 7 * np.cos(2 * t)

# 繪製填滿顏色的心形
plt.figure(figsize=(8, 8))
plt.fill(x, y, color='Pink')
plt.title('Heart Shape')
plt.gca().set_aspect('equal')  # 保持圖形比例
plt.axis('off')  # 隱藏座標軸
plt.show()
```
