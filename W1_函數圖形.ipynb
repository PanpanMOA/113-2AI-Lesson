#作業1 畫出一個函數圖形

```python
import numpy as np
import matplotlib.pyplot as plt

# 定義參數方程式
t = np.linspace(0, 2 * np.pi, 1000)
x = 16 * np.sin(t) ** 3
y = 8 * np.cos(t) - 7 * np.cos(2 * t)

# 繪製填滿顏色的心形
plt.figure(figsize=(8, 8))
plt.fill(x, y, color='Pink')
plt.title('Heart Shape')
plt.gca().set_aspect('equal')  # 保持圖形比例
plt.axis('off')  # 隱藏座標軸
plt.show()
```
